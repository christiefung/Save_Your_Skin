# Donwload Images
- Based on the review+image .csv file obtained in the main_amazon_scraper_2.0.ipynb, which is stored in the data/ folder, donwload the images using the urls. Note that each csv should represent one class (oily skin lotion, dry skin lotion, normal skin lotion)

```python
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook as tqdm
from datetime import datetime

import pandas as pd
import requests
import pickle
```

# Load the review+image csv

```python
# parameter (WARNING: make sure to change the fn and skin_type)!!!!!!!!!!!!!!
fn = '04-06-2020_11-22-48_amazon_review_with_normal_skin.csv'
skin_type = 'normal_skin' # used to name the folder saved in the data/ folder for all the images

# ------------------------- start
# Load a csv related to a skin class
df = pd.read_csv('../data/%s' %fn)
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

# Create an empty folder to store the image
-  create a new folder in /data/[skin_type]/, the folder will be used to stores images

```python
import os
folder_path = '../data/%s' %skin_type
try:
    os.mkdir(folder_path)
except:
    pass

# print
print('The images will be stored here:', folder_path)
```

# Loop the dataframe and download the images into the desired folder

```python
from tqdm import tqdm_notebook as tqdm
import urllib.request
import time
```

```python
from tqdm import tqdm_notebook as tqdm
```

```python
df.shape[0]
```

```python
for i in tqdm(range(0, (df.shape[0]))):
    df_tmp = df.iloc[i]
    
    # get the id for the images
    asin = df_tmp['ASIN']
    img_pid = df_tmp['pid'] #unique id in the csv corresponding to a row of review and image links
    
    # get the img link; note that there can be more than 1 image link
    # because a reivew can have more than 1 image
    # split using ','
    img_links = df_tmp['img_link']
    img_links = img_links.split(',')
    
    # for each image link, download it and save it into /data/[skin_type]/ folder
    # file name of an image: [skin_type]_[asin]_pid[img_pid]_[count], where [count] is the No. of images
    # corresponding to an img_pid
    count = 0
    print('Downloading images......')
    for link in img_links:
        img_file_name = '%s_%s_pid%s_%s' %(skin_type, asin, img_pid, count)
        try:
            print(link)
            urllib.request.urlretrieve(link, "%s/%s.jpg" %(folder_path, img_file_name))
            count += 1
            
            # wait for 0.5 second to make sure I download the image successfully before the next one
            time.sleep(0.5)
        except:
            pass
        
```
